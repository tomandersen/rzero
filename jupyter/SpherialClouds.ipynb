```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

```

```python
# d - distance between centre of sphere. x1, etc position relative to 
def calcForce(d, x1, y1, z1, x2, y2, z2, r1, r2):
    if (x1*x1 + y1*y1 + z1*z1) > r1*r1:
        return 0
    if (x2*x2 + y2*y2 + z2*z2) > r2*r2:
        return 0

    # calculate force, only return x component. Should work with symmetry
    # maybe return vector
    netx =  d - x1 + x2
    nety =  d - y1 + y2
    netz =  d - z1 + z2
    
    dSq = netx*netx + nety*nety + netz*netz

    # x component:
    # mult by unit vector in x direction.
    #force = np.sqrt(netx*netx/dSq)
    force = np.sqrt(netx*netx/dSq)/dSq
    if netx < 0: 
        force = -force
    
    return force
```

```python
r1 = 20.0 # sphere 1 radius
r2 = 20.0 # sphere 2 radius
d1 = 50.456456

numGrid = 3
r1Step = r1/numGrid
r2Step = r2/numGrid

forces = []
distances = []
maxForce = 0

#calcForce = np.frompyfunc(calcForce, 9, 1)


for d in np.arange(0.34678, d1, 0.987643):
    forceSum = 0.0
    for x1 in np.arange(-r1, r1, r1Step):
        for y1 in np.arange(-r1, r1, r1Step):
            for z1 in np.arange(-r1 + r1Step*0.5, r1, r1Step):
                for x2 in np.arange(-r2, r2, r2Step):
                    for y2 in np.arange(-r2, r2, r2Step):
                        for z2 in np.arange(-r2, r2, r2Step):
                            forceSum = forceSum + calcForce(d, x1, y1, z1, x2, y2, z2, r1, r2)
    maxForce = max(forceSum, maxForce)
    print('d:',d)
    #print('force:',forceSum)
    distances.append(d)
    forces.append(forceSum)

#print('forces:',forces)

fig = plt.figure(figsize=(7, 5), dpi=200)
plt.plot(distances, forces, marker="none", color='red', lw=1.5, label='Force')
plt.xlabel('distance');
plt.ylabel('force');

plt.legend(loc='lower right')
plt.axis((0, d1, 0.0, maxForce));
#plt.savefig('../chapters/images/force.png', bbox_inches='tight')
plt.show()
```
